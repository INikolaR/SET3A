### Функции для рисования

Импортируем matplotlib

```python
import matplotlib.pyplot as plt
```

Функция рисования нескольких графиков на одной диаграмме

```python
def make_plot(list_of_input_files, title):
    plt.rcParams["figure.figsize"] = (10, 10)
    to_legend = []
    for file in list_of_input_files:
        x = []
        y = []
        with open(file[0]) as f:
            for line in f:
                pair = list(map(int, line.split()))
                x.append(pair[0])
                y.append(pair[1])
        graph, = plt.plot(x, y, color=file[1], label=file[2])
        plt.plot(x, y, ".", color=file[1])
        to_legend.append(graph)
    plt.legend(handles=to_legend)
    plt.title(title)
```

### Графики замеров времени

### Выводы

В некоторых алгоритмах во избежания копирования строк вместо сортировки строк сортировались указатели на строки.

Замеры времени в программах проводились несколько раз и результат усреднялся.

Так как все алгоритмы тестировались на одинаковых наборах строк, сделанным тестированием не получится показать зависимость асимптотики от средней длины строк. Графики показывают только зависимость времени и количества сравнений от числа строк $n$.

#### Временные асимптотики

Как можно видеть, временная асимптотика квадратичная (видны отчётливые параболы) у всех алгоритмов, кроме msd radix sort без переключения на string quicksort, у которого она экспоненциальная. Экспоненциальная асимптотика у radix + quick sort наблюдается из-за большого количества веток рекурсии, вызов которых занимает много времени, в то время как сами ветки делают мало сравнений. Квадратичное время работы - вероятно, из-за проблем с конкретной реализацией. Алгоритмы реализовывались в точности так, как описано в лекциях/дополнениях к лекциям (я полагаю, именно такие реализацции от нас и ожидались), и в этих реализациях есть много копирований элементов (хоть это и указатели на строки - на копирование большого их количества тоже нужно время). 

#### Асимптотики сравнений

У radix sort асимптотика линейная (хоть и с большей константой, чем у других алгоритмов). У radix sort + quick sort - также линейная, но уже с меньшей константой. У остальных алгоритмов асимптотикаа $n \log n$, хотя график такой функции и не отличим от линейного на соответствующих размерах данных. Это согласуется с теоретическими асимптотиками. Также у графиков quick sort и string quick sort видны явные колебания прямой вверх-вниз. Это объясняется случайным выбором опорного элемента: из-за случайности количество сравнений может быть немного больше или меньше среднего.
